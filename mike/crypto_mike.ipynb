```python
## Import Dependencies
import yfinance as yf
import pandas as pd
```

```python
## Ticker List and Start date
tickers = ["BTC-USD", "ETH-USD", "^CMC200", "COIN", "GLD", "SLV", "^VIX", "CL=F", "NQ=F", "ES=F", "^SPSIBK", "EURUSD=X", "^TNX"]
start_date = "2019-12-01"
```

```python
## Look up ticker and create dataset
for ticker in tickers:
    data = pd.DataFrame(yf.download(ticker, start=start_date))
    filename = ticker + '.csv'
    data.to_csv(filename)
```
