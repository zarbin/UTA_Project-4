```python
#import dependencies
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
## Ticker List and Start date
## why take out VIX and Banking Sector?
tickers = ["BTC-USD", "ETH-USD", "BITW", "GLD", "SLV", "CL=F", "SVOL", "^IXIC", "^GSPC", "^DJI", "META", "AMZN", "AAPL", "NFLX", "GOOG", "TSLA", "XLF", "UUP", "IEF"]
start_date = "2019-12-01"

# Fetch historical data
market_data = yf.download(tickers, start=start_date)

# Dataframe preview
market_data.head()
```

```python
df = pd.DataFrame()


for stock in tickers: 
    info = yf.Ticker(stock).info
    print(info)
    marketcap = info.get('marketCap', None)
    shortname = info.get('shortName', None)   
    longname = info.get('longName', None)    
    df = df.append({'Stock':stock,'marketcap':marketcap, 'shortname':shortname}, ignore_index=True)
    
print(df)
```

```python
# Add custom labels to the DataFrame based on the ticker symbol
labels = ['Crypto', 'Crypto', 'Crypto', 'Commodities', 'Commodities', 'Commodities', 'Volatility', 'Indexes', 'Indexes', 'Indexes', 'Tech', 'Tech', 'Tech', 'Tech', 'Tech', 'Tech', 'Finance', 'Currency', 'Bonds']
df['Label'] = labels[:len(df)]
```
