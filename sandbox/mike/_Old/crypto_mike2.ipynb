```python
## Import Dependencies
import yfinance as yf
import pandas as pd
from datetime import date, timedelta, datetime
```

```python
## Ticker List, start date
tickers = ["BTC-USD", "ETH-USD", "^CMC200", "GLD", "SLV", "^VIX", "CL=F", "NQ=F", "ES=F", "XLF", "EURUSD=X", "^TNX"]
start_date = "2019-10-31"
```

```python
## Interval lengths
long = 30
med = 15
short = 5
```

```python
## Look up ticker and create csv
for ticker in tickers:
    data = pd.DataFrame(yf.download(ticker, start=start_date))
    filename = ticker + '.csv'
    data.to_csv(filename)
```

```python
## Create comprehensive list of dates
start = datetime.strptime(start_date, '%Y-%m-%d').date()
end = date.today() - timedelta(days = 1)
delta = timedelta(days=1)

dates = []
while start <= end:
    dates.append(start.isoformat())
    start += delta

dates = pd.DataFrame(dates)
dates.columns =['Date']
dates['Date'] = pd.to_datetime(dates['Date'])
dates
```

```python
## Upload csv datasets
gld = pd.read_csv(tickers[0] + '.csv')
gld['Date'] = pd.to_datetime(gld['Date'])
gld
```

```python
## Merge dfs to capture all dates including weekends and market holidays
test = pd.merge(dates, gld, how = 'left', on='Date')
test
```

```python
## Fill in weekends and market holidays using last trading day
test['Open'].fillna(method='ffill', inplace=True)
test['High'].fillna(method='ffill', inplace=True)
test['Low'].fillna(method='ffill', inplace=True)
test['Close'].fillna(method='ffill', inplace=True)
test['Adj Close'].fillna(method='ffill', inplace=True)
test['Volume'].fillna(method='ffill', inplace=True)
test
```

```python
# Calculate 'Adj Close' for past and future
test['hi_shift'] = test['High'].shift(1)
test['lo_shift'] = test['Low'].shift(1)
test['adjclose_shift'] = test['Adj Close'].shift(1)
test['vol_shift'] = test['Volume'].shift(1)

test['close_last_long'] = test['Adj Close'].shift(long)
test['close_last_med'] = test['Adj Close'].shift(med)
test['close_last_short'] = test['Adj Close'].shift(short)
test['close_last_1'] = test['Adj Close'].shift(2)
test['close_next_1'] = test['Adj Close']
test['close_next_short'] = test['Adj Close'].shift(-short + 1)
test['close_next_med'] = test['Adj Close'].shift(-med + 1)
test['close_next_long'] = test['Adj Close'].shift(-long + 1)
test
```

```python
test.to_csv('test2.csv')
```

```python
## Calculate short, med, long hi/lo, avg vol, return
# Past
test['avgvol_last_long'] = test['vol_shift'].rolling(long).sum() / long
test['hi_to_lo_last_long'] = (test['adjclose_shift'].rolling(long).max() / test['adjclose_shift'].rolling(long).min()) - 1
test['return_last_long'] = (test['adjclose_shift'] / test['close_last_long']) - 1

test['avgvol_last_med'] = test['vol_shift'].rolling(med).sum() / med
test['hi_to_lo_last_med'] = (test['adjclose_shift'].rolling(med).max() / test['adjclose_shift'].rolling(med).min()) - 1
test['return_last_med'] = (test['adjclose_shift'] / test['close_last_med']) - 1

test['avgvol_last_short'] = test['vol_shift'].rolling(short).sum() / short
test['hi_to_lo_last_short'] = (test['adjclose_shift'].rolling(short).max() / test['adjclose_shift'].rolling(short).min()) - 1
test['return_last_short'] = (test['adjclose_shift'] / test['close_last_short']) - 1

test['hi_to_lo_last_1'] = (test['hi_shift'] / test['lo_shift']) - 1
test['return_last_1'] = (test['adjclose_shift'] / test['close_last_1']) - 1

# Future
test['return_next_1'] = (test['close_next_1'] / test['adjclose_shift']) - 1
test['return_next_short'] = (test['close_next_short'] / test['adjclose_shift']) - 1
test['return_next_med'] = (test['close_next_med'] / test['adjclose_shift']) - 1
test['return_next_long'] = (test['close_next_long'] / test['adjclose_shift']) - 1

test
```

```python
## Filter out Nulls
test = test.dropna()
test
```

```python
## Drop unused fields
test = test.drop(columns=['hi_shift', 'lo_shift', 'adjclose_shift',
                          'close_last_long', 'close_last_med', 'close_last_short', 'close_last_1', 'close_next_1', 'close_next_long', 'close_next_med', 'close_next_short'])
test
```

```python
## Rename and rearrange columns
# Rename
test = test.rename(columns={"vol_shift": "avgvol_last_1",
                            "avgvol_last_long": "avgvol_last_" + str(long), 
                            "hi_to_lo_last_long": "hi_to_lo_last_" + str(long),
                            "return_last_long": "return_last_" + str(long),
                            "avgvol_last_med": "avgvol_last_" + str(med), 
                            "hi_to_lo_last_med": "hi_to_lo_last_" + str(med),
                            "return_last_med": "return_last_" + str(med),
                            "avgvol_last_short": "avgvol_last_" + str(short), 
                            "hi_to_lo_last_short": "hi_to_lo_last_" + str(short),
                            "return_last_short": "return_last_" + str(short),
                            "return_next_long": "return_next_" + str(long),
                            "return_next_med": "return_next_" + str(med),
                            "return_next_short": "return_next_" + str(short)})

# Rearrange
test = test[['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 
             'avgvol_last_' + str(long), 'hi_to_lo_last_' + str(long), 'return_last_' + str(long),
             'avgvol_last_' + str(med), 'hi_to_lo_last_' + str(med), 'return_last_' + str(med),
             'avgvol_last_' + str(short), 'hi_to_lo_last_' + str(short), 'return_last_' + str(short),
             'avgvol_last_1', 'hi_to_lo_last_1', 'return_last_1',
             'return_next_1', 'return_next_' + str(short), 'return_next_' + str(med),'return_next_' + str(long)]]
test
```
