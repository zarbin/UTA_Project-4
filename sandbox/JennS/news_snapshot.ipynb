```python
import pandas as pd
from newsapi import NewsApiClient

# Import api key
from config import api_key_news
```

# Extract News

```python
# Start session
newsapi = NewsApiClient(api_key=api_key_news)
```

```python
crypto_snapshot = newsapi.get_everything(q='bitcoin OR crypto OR doge',
                                      from_param='2023-03-11',
                                      to='2023-04-09',
                                      language='en')
```

```python
crypto_snapshot
```

```python
bank_snapshot = newsapi.get_everything(q='bank',
                                      from_param='2023-03-11',
                                      to='2023-04-09',
                                      language='en')
```

```python
bank_snapshot
```

# Transform
Transform into dataframes to load as csv's

### Crypto

```python
crypto_articles = crypto_snapshot['articles']
```

```python
crypto_df = pd.DataFrame(crypto_articles)
```

```python
crypto_snapshot_df = crypto_df.loc[:,['publishedAt', 'title', 'description']]
crypto_snapshot_df.rename(columns={'publishedAt': 'date'}, inplace = True)

```

```python
crypto_snapshot_df['summary'] = crypto_snapshot_df['title'] + crypto_snapshot_df['description']
```

```python
crypto_snapshot_df.drop(columns=['title','description'], inplace=True)
crypto_snapshot_df
```

### Banking News

```python
banking_articles = bank_snapshot['articles']
banking_df = pd.DataFrame(banking_articles)

banking_snapshot_df = banking_df.loc[:,['publishedAt', 'title', 'description']]
banking_snapshot_df.rename(columns={'publishedAt': 'date'}, inplace = True)
```

```python
banking_snapshot_df['summary'] = banking_snapshot_df['title'] + banking_snapshot_df['description']
```

```python

banking_snapshot_df.drop(columns=['title','description'], inplace=True)
banking_snapshot_df
```
